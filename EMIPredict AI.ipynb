## Bank_balance

```python
#Convert to numeric, then fill with median

# Clean the bank_balance column properly
df['bank_balance'] = (
    df['bank_balance']
    .astype(str)                              # ensure all entries are strings
    .str.replace(r'[‚Çπ$,]', '', regex=True)    # remove ‚Çπ, $, and commas
    .str.replace(r'\s+', '', regex=True)      # remove any spaces
    .str.replace(r'\.{2,}', '.', regex=True)  # replace multiple dots with one
)

# Handle bad values like '270700.0.0' safely
def safe_to_float(x):
    try:
        return float(x)
    except:
        return None

df['bank_balance'] = df['bank_balance'].apply(safe_to_float)

# Fill missing values with median
df['bank_balance'].fillna(df['bank_balance'].median(), inplace=True)

# Confirm conversion
print(df['bank_balance'].head())
print(df['bank_balance'].dtype)
```

```python
import numpy as np


# Select numeric columns
numeric_cols = df_train_capped.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Apply IQR capping and replace original columns
for col in numeric_cols:
    Q1 = df_train_capped[col].quantile(0.25)
    Q3 = df_train_capped[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    
    # Cap in place (replace original column)
    df_train_capped[col] = np.where(
        df_train_capped[col] > upper_limit, upper_limit,
        np.where(df_train_capped[col] < lower_limit, lower_limit, df_train_capped[col])
    )
    
    print(f"{col}: lower={lower_limit}, upper={upper_limit}")

# Verify
df_train_capped.head()
```

## Feature Engineering for training and testing data

```python
import numpy as np
import pandas as pd

def feature_engineering(df):
    df = df.copy()

    # --- Derived Ratios ---
    df['debt_to_income_ratio'] = df['current_emi_amount'] / df['monthly_salary']
    df['expense_to_income_ratio'] = (
        df['monthly_rent'] + df['groceries_utilities'] + df['other_monthly_expenses']
    ) / df['monthly_salary']
    df['affordability_ratio'] = df['max_monthly_emi'] / df['monthly_salary']
    df['credit_utilization_ratio'] = df['requested_amount'] / (df['bank_balance'] + 1)  # +1 to avoid /0

    # --- Risk / Stability Features ---
    df['employment_stability_score'] = np.log1p(df['years_of_employment']) * (df['credit_score'] / 100)
    df['financial_stress_index'] = (
        df['debt_to_income_ratio'] + df['expense_to_income_ratio'] + df['credit_utilization_ratio']
    )

    # --- Optional: replace infinities and NaN ---
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df.fillna(0, inplace=True)

    return df


# Apply on both train and test encoded data
df_train_engineered = feature_engineering(df_train_encoded)
df_test_engineered = feature_engineering(df_test_encoded)

print("‚úÖ Feature engineering completed successfully!")
print("Train shape:", df_train_engineered.shape)
print("Test shape:", df_test_engineered.shape)
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import joblib
import pandas as pd


# -----------------------------
# Load Engineered Datasets
# -----------------------------
df_train_engineered = pd.read_csv('emi_prediction_dataset_train_engineered.csv')
df_test_engineered = pd.read_csv('emi_prediction_dataset_test_engineered.csv')

# ------------------------------
# Train & Test Split
# ------------------------------
X_train_reg = df_train_engineered.drop('max_monthly_emi', axis=1)
y_train_reg = df_train_engineered['max_monthly_emi']
X_test_reg = df_test_engineered.drop('max_monthly_emi', axis=1)
y_test_reg = df_test_engineered['max_monthly_emi']




# Initialize XGBoost Regressor
xgb_reg = xgb.XGBRegressor(
    n_estimators=200,       # Number of boosting rounds/trees
    learning_rate=0.05,     # Step size shrinkage
    max_depth=6,            # Maximum tree depth
    subsample=0.8,          # Row subsampling
    colsample_bytree=0.8,   # Column subsampling
    random_state=42,
    n_jobs=-1
)

# Train the model
xgb_reg.fit(X_train_reg, y_train_reg)



# Predict on test set
y_pred_reg = xgb_reg.predict(X_test_reg)

# Evaluate model
mse = mean_squared_error(y_test_reg, y_pred_reg)
mae = mean_absolute_error(y_test_reg, y_pred_reg)
r2 = r2_score(y_test_reg, y_pred_reg)

print("üìä XGBoost Regressor Evaluation:")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"R¬≤ Score: {r2:.2f}")

# Save model
joblib.dump(xgb_reg, "xgb_regressor_model.joblib")
print("‚úÖ Model saved as xgb_regressor_model.joblib")

```
